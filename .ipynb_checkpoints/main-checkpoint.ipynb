```python
import numpy as np
import matplotlib.pyplot as plt

def generate_vertical_branch(x, y, length, angle, depth):
    if depth == 0:
        return
    
    x_end = x + length * np.cos(angle)
    y_end = y + length * np.sin(angle)
    
    plt.plot([x, x_end], [y, y_end], color='black', linewidth=1)
    
    new_length = length * 0.7
    new_depth = depth - 1
    
    # Generate random angles for horizontal branching
    left_angle = angle + np.random.uniform(-np.pi/4, np.pi/4)
    right_angle = angle + np.pi + np.random.uniform(-np.pi/4, np.pi/4)
    
    generate_vertical_branch(x_end, y_end, new_length, left_angle, new_depth)
    generate_vertical_branch(x_end, y_end, new_length, right_angle, new_depth)

def plot_air_shower(density):
    # Generate random point for the air shower's vertical direction
    x_vertical = np.random.uniform(0, 100)
    
    # Plot the vertical line
    plt.plot([x_vertical, x_vertical], [0, 100], color='blue', linestyle='--', linewidth=1)
    
    # Generate branching pattern from the vertical line
    generate_vertical_branch(x_vertical, 50, 20, np.pi/2, density)
    
    plt.gca().set_aspect('equal', adjustable='box')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Extensive Air Shower')
    plt.show()

# Set the density of branching (adjust as desired)
density = 10
plot_air_shower(density)
```
